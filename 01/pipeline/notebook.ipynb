```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

```python
def load_table_to_db(df: pd.DataFrame, table_name: str):
    df.head(0).to_sql(
        name=table_name,
        con=engine,
        if_exists="replace"
    )
    print(f"{table_name} created")
    
    df.to_sql(
        name=table_name,
        con=engine,
        if_exists="append"
    )
    print(f"Table {table_name} loaded")
```

```python
# Read the green taxi trips data for November 2025 - logs
df_green_trips = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet")
df_green_trips.dtypes
```

```python
# Read zones - referential
df_zones = pd.read_csv('https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv')
df_zones.dtypes
```

```python
load_table_to_db(df_green_trips, "green_trips")
load_table_to_db(df_zones, "zones")
```
